引入模块

```python
from langchain.agents import create_agent
from langchain_community.chat_models.tongyi import ChatTongyi
from langchain_core.tools import tool
```

使用tool装饰器来创建工具

```python
@tool(description="查询天气")
def get_weather() -> str:
    return "晴天"
```

阅读create_agent方法源码，发现agent实例对象也是Runnable子类

```python
agent = create_agent(
    model=ChatTongyi(model="qwen3-max"),        # 智能体的大脑LLM
    tools=[get_weather],            # 向智能体提供工具列表
    system_prompt="你是一个聊天助手，可以回答用户问题。",
)
```

agent使用invoke方法传入的信息是一个字典dict

messages对应的是个列表，列表元素是messages类型的数据

```python
res = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": "明天深圳的天气如何？"},
        ]
    }
)
```

```python
print(res)
print(type(res["messages"][0]))
```

```python
for msg in res["messages"]:
    print(type(msg).__name__, msg.content)
```
